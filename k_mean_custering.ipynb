```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
X=data.iloc[:,[3,4]].values
```

now importing values

```python
from sklearn.cluster import KMeans
```

```python
wcs=[]
for i in range(1,11):
  kmean=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmean.fit(X)# now fit the model
  wcs.append(kmean.inertia_)
plt.plot(range(1,11),wcs)
plt.title("elbo")
plt.xlabel("number of cluster")
plt.ylabel(wcs)
plt.show()
```

we look where are grap is stablised if is is stablished we note the point..like i n this particulare graph the value is stablised at 5


```python
kmean=KMeans(n_clusters=i,init='k-means++',random_state=42)
y_label=kmean.fit_predict(X)
```

```python
plt.scatter(X[y_label==0,0],X[y_label == 0,1],s=100,c='red',label='cluster-1')
plt.scatter(X[y_label==1,0],X[y_label == 1,1], s=100,c='Yellow',label='cluster-2')
plt.scatter(X[y_label==2,0],X[y_label == 2,1],s=100,c='green',label='cluster-3')
plt.scatter(X[y_label==3,0],X[y_label == 3,1],s=100,c='pink',label='cluster-4')
plt.scatter(X[y_label==4,0],X[y_label == 4,1],s=100,c='blue',label='cluster-5')
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],s=200,c='black',label='center cluster')
plt.title("coustumer detaile")
plt.xlabel('income')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
